```python
#7question
fruits = ['apple','orange','mango','kiwi','cherry']
fruits2 = []
print(fruits)
for i in range(1,len(fruits)+1):
    fruits2.append(fruits[-i])
fruits=list(fruits2)
print(fruits)
```

```python
#8question
numbers = [91,60,70,87,77,55]
print(numbers)
numbers.sort()
print(f"ascending:\n{numbers}")
numbers.sort(reverse=True)
print(f"descending:\n{numbers}")
```

```python
#9question
numbers = [91,60,70,87,77,55,77,60]
numbers.sort()
final={}
times=1
for i in numbers:
   if i not in final.keys():
      final.update({i:f"{times} Times Repeated"})
   else:
      final.update({i:f"{times+1} Times Repeated"})
for i,j in final.items():
    print(i,j)
```
